 ### Part of Speech Tagging
 **Source:** https://youtu.be/6j6M2MtEqi8
 * **POS Tag list:** https://pythonprogramming.net/part-of-speech-tagging-nltk-tutorial/

* CC	coordinating conjunction
* CD	cardinal digit
* DT	determiner
* EX	existential there (like: "there is" ... think of it like "there exists")
* FW	foreign word
* IN	preposition/subordinating conjunction
* JJ	adjective	'big'
* JJR	adjective, comparative	'bigger'
* JJS	adjective, superlative	'biggest'
* LS	list marker	1)
* MD	modal	could, will
* NN	noun, singular 'desk'
* NNS	noun plural	'desks'
* NNP	proper noun, singular	'Harrison'
* NNPS	proper noun, plural	'Americans'
* PDT	predeterminer	'all the kids'
* POS	possessive ending	parent\'s
* PRP	personal pronoun	I, he, she
* PRP2 	possessive pronoun	my, his, hers
* RB	adverb	very, silently,
* RBR	adverb, comparative	better
* RBS	adverb, superlative	best
* RP	particle	give up
* TO	to	go 'to' the store.
* UH	interjection	errrrrrrrm
* VB	verb, base form	take
* VBD	verb, past tense	took
* VBG	verb, gerund/present participle	taking
* VBN	verb, past participle	taken
* VBP	verb, sing. present, non-3d	take
* VBZ	verb, 3rd person sing. present	takes
* WDT	wh-determiner	which
* WP	wh-pronoun	who, what
* WP$	possessive wh-pronoun	whose
* WRB	wh-abverb	where, when

```python
# Function 03
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            
            chunkGram = r"""Chunk: {<.*>+}
                                   }<VB.?|IN|DT|TO>+{"""
            
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            
            chunked.draw()  # Draw a parse trees
    except Exception as e:
        print(str(e))
```

## Part 12

### Words as Features for Learning

**Source:** https://youtu.be/-vVskDsHcVc

```python
import nltk
from nltk.corpus import movie_reviews
import random

documents = []

# List of positive & negetive comments
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((list(movie_reviews.words(fileid)), category))
        
# Shuffling or randomize
random.shuffle(documents)

all_words = []

for w in movie_reviews.words():
      w.lower() # convert into lowercase 
      all_words.append(w.lower())

all_words = nltk.FreqDist(all_words)
```

```python
print(all_words.most_common(20)) # First 20 most common words or charecters
```

```python
word_features = list(all_words.keys())[:3000]  # Top 3000 Words
```

```python
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
        
    return features
```

```python
print((find_features(movie_reviews.words('neg/cv000_29416.txt'))))
```

```python
featuresets = [(find_features(rev),category) for (rev, category) in documents]
```

## Part 13
### Naive Bayes Classification Algorithm
**Source:** https://youtu.be/rISOsUaTrO4

**Algorithm----**
 * https://www.analyticsvidhya.com/blog/2017/09/naive-bayes-explained/
 * https://www.geeksforgeeks.org/naive-bayes-classifiers/

```python
featuresets = [(find_features(rev),category) for (rev, category) in documents]
```

```python
train_set = featuresets[:1900] # First 1900 data are for training
test_set = featuresets[1900:] #  Last 1900 data are for testing
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set) # Training  
```

```python
print(f"Naive Bayes Classifier Accuracy:{(nltk.classify.accuracy(classifier, test_set))*100:.2f}%")
```

```python
classifier.show_most_informative_features(20)
```

## Part 16
### Combining Algos with a Vote
**Source:** https://youtu.be/vlTQLb_a564

```python
from nltk.classify import ClassifierI
from statistics import mode
```

```python
class VoteClassifier(ClassifierI):
    def __init__(self, *classifiers):
        self._classifiers = classifiers

    def classify(self, features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        return mode(votes)

    def confidence(self, features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)

        choice_votes = votes.count(mode(votes))
        conf = choice_votes / len(votes)
        return conf
```

```python
voted_classifier = VoteClassifier(classifier,
                                  NuSVC_classifier,
                                  LinearSVC_classifier,
                                  SVC_classifier,
                                  SGDClassifier_classifier,
                                  MNB_classifier,
                                  BRB_classifier,
                                  LogisticRegression_classifier)

print("voted_classifier accuracy percent:", (nltk.classify.accuracy(voted_classifier, test_set))*100)

print("Classification:", voted_classifier.classify(test_set[0][0]), "Confidence %:",voted_classifier.confidence(test_set[0][0])*100)
print("Classification:", voted_classifier.classify(test_set[1][0]), "Confidence %:",voted_classifier.confidence(test_set[1][0])*100)
print("Classification:", voted_classifier.classify(test_set[2][0]), "Confidence %:",voted_classifier.confidence(test_set[2][0])*100)
print("Classification:", voted_classifier.classify(test_set[3][0]), "Confidence %:",voted_classifier.confidence(test_set[3][0])*100)
print("Classification:", voted_classifier.classify(test_set[4][0]), "Confidence %:",voted_classifier.confidence(test_set[4][0])*100)
print("Classification:", voted_classifier.classify(test_set[5][0]), "Confidence %:",voted_classifier.confidence(test_set[5][0])*100)
print("Classification:", voted_classifier.classify(test_set[6][0]), "Confidence %:",voted_classifier.confidence(test_set[6][0])*100)
```
